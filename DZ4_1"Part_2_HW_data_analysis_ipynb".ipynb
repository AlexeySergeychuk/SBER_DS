```
def lastworkday(year, month):
  from calendar import monthcalendar
  lastday = monthcalendar(year, month)
  count = 0
  for i in range(5):
    if lastday[len(lastday)-1][i] != 0:
      count += 1
    else:
      pass

  print(f'Последний рабочий день месяца: {lastday[len(lastday)-1][count-1]}')
```
